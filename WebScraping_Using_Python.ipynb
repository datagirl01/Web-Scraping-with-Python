Example to obtain data from a webpage using 'get' command

```python
url = "https://api.namara.io/v0/data_sets/5d4c346d-325f-4a5f-9c05-f29f6342e8db/data/en-2?geometry_format=wkt&api_key=b26991aee69faf1f508226e7ca20c4266036254fa3a2703d18a18b15eccc5f91&organization_id=5c3f726fc9abee5de5895638&where=date_received%20%3E%20%222014-01-01%22&select=date_received%2Cproduct%2Cissue%2Csub_issue%2Ccompany_public_response%2Ccompany%2Cstate%2Czip_code"
```

# Extract data from a web page and write the data as csv file

```python
import csv
import os.path
with open('dict.csv', 'a') as csv_file:
    headers = data[0].keys()
    writer = csv.DictWriter(csv_file, headers)
    
    if(os.path.isfile('dict.csv')):
        writer.writeheader()
    
    writer.writerows(data)
```
